```python
# Install required libraries
!pip install --quiet opencv-python-headless matplotlib numpy scikit-image

import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import data

# Load a sample 8-bit grayscale image (Cameraman)
img = data.camera()   # Already 8-bit grayscale

# (a) Brightness Improvement
bright_img = cv2.convertScaleAbs(img, alpha=1, beta=50)

# (b) Brightness Reduction
dark_img = cv2.convertScaleAbs(img, alpha=1, beta=-50)

# (c) Thresholding
_, thresh_img = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

# (d) Negative of an image
neg_img = 255 - img

# (e) Log Transformation
c = 255 / np.log(1 + np.max(img))
log_img = c * (np.log(1 + img))
log_img = np.array(log_img, dtype=np.uint8)

# (f) Power-Law (Gamma) Transformation
gamma = 0.5   # try values like 0.4, 0.7, 1.5
gamma_img = np.array(255 * (img / 255) ** gamma, dtype=np.uint8)

# Show results
titles = ['Original', 'Brightened', 'Darkened', 'Threshold', 'Negative', 'Log', 'Gamma']
images = [img, bright_img, dark_img, thresh_img, neg_img, log_img, gamma_img]

plt.figure(figsize=(15, 10))
for i in range(len(images)):
    plt.subplot(2, 4, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.show()
```
