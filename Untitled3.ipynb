```python
#  Install libraries
!pip install --quiet opencv-python-headless matplotlib numpy scikit-image

import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import data

# Step 1: Load an image (8-bit grayscale)
img = data.camera()  # Cameraman image

# Step 2: Plot original image and histogram
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.hist(img.ravel(), 256, [0,256], color='black')
plt.title("Histogram (Original)")
plt.show()

# Step 3: Histogram Equalization
equalized_img = cv2.equalizeHist(img)

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.imshow(equalized_img, cmap='gray')
plt.title("Histogram Equalized Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.hist(equalized_img.ravel(), 256, [0,256], color='black')
plt.title("Histogram (Equalized)")
plt.show()

# ------------------------------------------------
# Step 4: Spatial Domain Filters
# (a) Average Smoothing
avg_filtered = cv2.blur(img, (5,5))

# (b) Gaussian Smoothing
gauss_filtered = cv2.GaussianBlur(img, (5,5), 1)

# (c) Median Filtering
median_filtered = cv2.medianBlur(img, 5)

# (d) Laplacian Sharpening
laplacian = cv2.Laplacian(img, cv2.CV_64F)
laplacian = cv2.convertScaleAbs(laplacian)
sharpened = cv2.addWeighted(img, 1.5, laplacian, -0.5, 0)

# ------------------------------------------------
# Step 5: Frequency Domain Filters
# Convert image to frequency domain
f = np.fft.fft2(img)
fshift = np.fft.fftshift(f)

rows, cols = img.shape
crow, ccol = rows//2 , cols//2

# Low-pass filter (keep center)
mask_lp = np.zeros((rows, cols), np.uint8)
mask_lp[crow-30:crow+30, ccol-30:ccol+30] = 1
low_pass = fshift * mask_lp
low_img = np.abs(np.fft.ifft2(np.fft.ifftshift(low_pass)))

# High-pass filter (remove center)
mask_hp = 1 - mask_lp
high_pass = fshift * mask_hp
high_img = np.abs(np.fft.ifft2(np.fft.ifftshift(high_pass)))

# ------------------------------------------------
# Step 6: Show all results
titles = [
    "Original", "Equalized", "Average Blur", "Gaussian Blur",
    "Median Filter", "Sharpened", "Low-Pass (Freq)", "High-Pass (Freq)"
]
images = [img, equalized_img, avg_filtered, gauss_filtered,
          median_filtered, sharpened, low_img, high_img]

plt.figure(figsize=(15,10))
for i in range(len(images)):
    plt.subplot(2,4,i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.show()
```
